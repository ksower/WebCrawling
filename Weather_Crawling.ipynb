# 다운받은 데이터 압축 해제 및 DB 저장(황사기후)
사용하는 DB는 SQLite3 입니다.

```python
import os
import zipfile
import shutil

# 데이터가 위치한 폴더 및 해제할 폴더
before_folder = './dataset/crawling_result/'
extract_folder = './dataset/extract/'
after_folder = './dataset/weather_yellowdust/'

# 필요한 폴더들을 생성
def createFolder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(directory)
            print(directory, ' 폴더 생성 완료...')
    except OSError:
        print ('Error: Creating directory. ' +  directory)
createFolder(before_folder)
createFolder(extract_folder)
createFolder(after_folder)
print('\n')


# 압축해제할 상위 압축 폴더
zd1 = []
for file in os.listdir(before_folder):
    if file[-3:] == 'zip':
        zd1.append(before_folder + file)
# 상위 압축 폴더를 압축 해제
for file in zd1:
    zipfile.ZipFile(file).extractall(path=extract_folder)
print('상위 압축 폴더 해제 완료...')
    
#압축해제할 하위 압축 폴더
zd2 = []
for file in os.listdir(extract_folder):
    if file[-3:] == 'zip':
        zd2.append(extract_folder + file)
# 하위 압축 폴더를 압축 해제
for file in zd2:
    zipfile.ZipFile(file).extractall(path=after_folder)
print('하위 압축 폴더 해제 완료...')


# 필요없는 중간 폴더 삭제
if os.path.exists(extract_folder):
    shutil.rmtree(extract_folder)
print('중간 폴더 삭제 완료...')
print('프로세스 완료!')
```

```python
import os
import pandas as pd

after_folder = './dataset/weather_yellowdust/'

fcsv = []
for file in os.listdir(after_folder):
    fcsv.append(after_folder + file)
print('총 csv 파일 수 : ', len(fcsv))

all_data = []
data = pd.DataFrame(columns=['시간'])
last_loc = '0'
cnt = 0
for file in fcsv:
    df = pd.read_csv(file, encoding='cp949')
    
    cur_loc = df['지점'][0]
    df.rename(columns={'1시간평균 미세먼지농도(㎍/㎥)':cur_loc}, inplace=True)
    df.drop(['지점'], axis=1, inplace=True)
    
    if '일시' in df.columns:
        df.rename(columns={'일시':'시간'}, inplace=True)
    if last_loc == cur_loc:
        data = pd.merge(data, df, on=['시간', last_loc], how='outer')
    else:
        all_data.append(data)
        data = pd.DataFrame(columns=['시간'])
        data = pd.merge(data, df, on='시간', how='outer')
        last_loc = cur_loc
    
    cnt = cnt + 1
    print('(', cnt, ')  ', file, ' 파일 탐색 완료...')
all_data.append(data)
print('모든 파일 탐색 완료!\n\n')


print('데이터 병합 중...')
data = pd.DataFrame(columns=['시간'])
for per in all_data:
    data = pd.merge(data, per, on='시간', how='outer')
print('데이터 병합 완료!')
```

잘 병합되었는지 마지막 시간을 확인해봅니다.

DB에 저장하기 전 시간 순서대로 정렬을 합니다.

```python
data = data.sort_values('시간').reset_index(drop=True)
data
```

sqlite3에 정리된 데이터를 저장합니다. local_yellow_dust 테이블입니다.

```python
import sqlite3

conn = sqlite3.connect("yellow_dust.db")
data.to_sql('local_yellow_dust_table', conn, chunksize=1000)
conn.commit()
conn.close()
```
