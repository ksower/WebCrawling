```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time


# 기상청 페이지에서 황사 데이터를 가져오는 url
driver_path = '크롬 드라이버 경로'
URL = 'https://data.kma.go.kr/data/climate/selectDustRltmList.do?pgmNo=68&tabNo=1'
end_page_num = 45  # 마지막 페이지보다 하나 더 많게


# 개인정보 [주의 필요!]
login_email = '계정'
login_pwd = '비번'

driver = webdriver.Chrome(executable_path=driver_path)
driver.get(URL)

time.sleep(5)
# 로그인
driver.find_element_by_id('loginBtn').click()
driver.find_element_by_id('loginId').send_keys(login_email)
driver.find_element_by_id('passwordNo').send_keys(login_pwd)
driver.find_element_by_class_name('btn_login').click()

time.sleep(5)

# 첫 번재 페이지 크롤링
# 전체 파일 선택
check_all = driver.find_element_by_css_selector('div.wrap_tbl.scroll_x > table > thead > tr > th.col_chk')
check_all.click()
time.sleep(1)
# 다운로드 버튼 클릭
driver.find_element_by_class_name('btn_download').click()
time.sleep(1)
# 다운로드 버튼 클릭 시 뜨는 팝업창
driver.find_element_by_id('reqstPurposeCd7').click()
driver.find_element_by_class_name('btn_request').click()

time.sleep(30) # 다운로드 시간이 오래 걸려 넉넉히 30초 대기
print('첫 번째 페이지 다운로드 완료...')



# 페이지별로 크롤링
for i in range(2, end_page_num):
    # 페이지 이동(1, 11, 21 등의 1단위 페이지는 다음 페이지 버튼 클릭하여 페이지 변경)
    if i % 10 == 1:
        driver.find_element_by_link_text('다음 페이지').click()
    else:
        driver.find_element_by_link_text(str(i)).click()
    time.sleep(2)
    
    # 전체 파일 선택
    check_all = driver.find_element_by_css_selector('div.wrap_tbl.scroll_x > table > thead > tr > th.col_chk')
    check_all.click()
    time.sleep(1)
    # 다운로드 버튼 클릭
    driver.find_element_by_class_name('btn_download').click()
    time.sleep(1)
    # 다운로드 버튼 클릭 시 뜨는 팝업창
    driver.find_element_by_id('reqstPurposeCd7').click()
    driver.find_element_by_class_name('btn_request').click()
    
    time.sleep(30) # 다운로드 시간이 오래 걸려 넉넉히 30초 대기
    print(i, ' 번째 페이지 다운로드 완료...')

print('\n전체 페이지 크롤링 완료!')
print('드라이버 종료.')
driver.quit()
```
